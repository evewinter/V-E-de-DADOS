#Precisão e arredondamento de dados quantitativos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Carregando o dataset
from google.colab import drive
drive.mount('/content/drive')
dataset_path = "/content/drive/MyDrive/spotifydata.csv"
df = pd.read_csv(dataset_path)

# Pegando a coluna 'tempo(beats por minuto)' e arredondando
tempo_nao_arredondado = np.round(df['tempo'], 5)
tempo_arredondado = np.round(df['tempo'], 2)

print("\nDados NÃO arredondados:", tempo_nao_arredondado.head())
print("\nDados ARREDONDADOS:", tempo_arredondado.head())

# Criando gráficos
plt.figure(figsize=(12, 5))

# Gráfico de dados não arredondados
plt.subplot(1, 2, 1)
plt.hist(tempo_nao_arredondado.dropna(), bins=30, color='pink', edgecolor='black')
plt.title('tempo NÃO Arredondado')
plt.xlabel('tempo')
plt.ylabel('Frequência')

# Gráfico de dados arredondados
plt.subplot(1, 2, 2)
plt.hist(tempo_arredondado.dropna(), bins=30, color='purple', edgecolor='black')
plt.title('tempo ARREDONDADOS')
plt.xlabel('tempo')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```

Vsualização dos dados da coluna 'Tempo', que lista os beats por minuto.

Primeiro vemos sem arrendodamento, com 5 casas decimais depois da vírgula e depois com os dados arredondados para 2 casas decimais.

Uma grande diferença é na coluna 2, onde o número arredonda de 97.998 para 98.

# Representação gráfica e tabular de dados qualitativos.

```python
from google.colab import drive
drive.mount('/content/drive')
dataset_path = "/content/drive/MyDrive/spotifydata.csv"
df = pd.read_csv(dataset_path)

genres_coluna = df['genres']

tabela_frequencia = df['genres'].value_counts().reset_index()
tabela_frequencia.columns = ['genres', 'Frequência']
tabela_frequencia['Frequência Relativa'] = tabela_frequencia['Frequência'] / tabela_frequencia['Frequência'].sum()

print(tabela_frequencia)

# Gerando o gráfico de pizza
plt.figure(figsize=(10, 8))
plt.pie(
    tabela_frequencia['Frequência'],
    labels=tabela_frequencia['genres'],
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Paired(range(len(tabela_frequencia))),  # Cores automáticas
    wedgeprops={'edgecolor': 'black'}  # Borda preta para separar as fatias
)
plt.title('Distribuição dos Gêneros Musicais', fontsize=16)
plt.tight_layout()
plt.show()
```

Tabela de frequência da coluna "genres", com os genêros musicais do dataset. Também gerei um gráfico de pizza com a porcentagem da tabela, mostrando uma representação gráfica desses dados qualitativos.  

# Medidas descritivas de dados quantitativos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Montando o Google Drive e carregando o dataset
from google.colab import drive
drive.mount('/content/drive')
dataset_path = "/content/drive/MyDrive/spotifydata.csv"
df = pd.read_csv(dataset_path)

# Verificando se as colunas existem no dataset
if 'duration_ms' in df.columns and 'genres' in df.columns:
    # Removendo valores nulos
    df_clean = df[['duration_ms', 'genres']].dropna()

    # Configurando o tamanho e estilo do gráfico
    plt.figure(figsize=(12, 8))
    sns.boxplot(data=df_clean, x='genres', y='duration_ms', palette='Set3')

    # Personalizando o gráfico
    plt.title('Boxplot de Duration_ms por Genres', fontsize=16)
    plt.xlabel('Genres', fontsize=14)
    plt.ylabel('Duração (ms)', fontsize=14)
    plt.xticks(rotation=45)  # Rotacionando os gêneros para melhor visualização
    plt.tight_layout()
    plt.show()
else:
    print("As colunas 'duration_ms' e 'genres' não foram encontradas no dataset.")
```

O Bloxpot relaciona a coluna genres com a coluna duration_ms (duração em milisegundos). Observamos que o boxplot do gênero "Alternative Metal" é mais alto que o de "Social Media Pop", por exemplo. Isso indica que as músicas de "Alternative Metal" tendem a ter maior duração. Os outliers mostram isso também, há uma música outlier do genêro com quase 500000 milisegundos.

# Visualização de Associação entre Variáveis qualitativas e quantitativas.

```python
import pandas as pd
import seaborn as sns

# Montando o Google Drive e carregando o dataset
from google.colab import drive
drive.mount('/content/drive')
dataset_path = "/content/drive/MyDrive/spotifydata.csv"
df = pd.read_csv(dataset_path)

contingencia = pd.crosstab(df["artist_popularity"], df["genres"])
print(contingencia)
```

A tabela de contingência cruza a variável quantitativa "artist_popularity" que mede a popularidade das músicas de 1 à 100, com a variável qualitativa "genres". Por exemplo, na linha 65.0 e coluna afrobeats, o valor 3 indica que existem 3 músicas de afrobeats com popularidade 65.

#Gráfico de dispersão: associação entre variáveis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Montando o Google Drive e carregando o dataset
from google.colab import drive
drive.mount('/content/drive')
dataset_path = "/content/drive/MyDrive/spotifydata.csv"
df = pd.read_csv(dataset_path)

# Criando o gráfico de dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='energy', y='loudness', alpha=0.6, color='purple')
plt.title('Relação entre Energia e Volume das Músicas')
plt.xlabel('Energia')
plt.ylabel('Volume (dB)')
plt.grid(True)
plt.show()
```

O gráfico tende a crescer, em uma correlação positiva, pois músicas com maior energia tendem a ter maior volume.
